```python
from langchain_community.tools import WikipediaQueryRun
from langchain_community.utilities import WikipediaAPIWrapper
from langchain_chroma import Chroma
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_core.documents import Document
from langchain_community.tools import DuckDuckGoSearchRun
from langchain_community.utilities import DuckDuckGoSearchAPIWrapper

```

```python
api_wraper = DuckDuckGoSearchAPIWrapper(max_results=10)
search = DuckDuckGoSearchRun(api_wrapper= api_wraper)
text = search.run('Egyptian Medical')
```

```python
text
```

```python
embedding = HuggingFaceEmbeddings(model = 'sentence-transformers/all-MiniLM-L6-v2')
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
doc = Document(page_content=text)
spiltter = RecursiveCharacterTextSplitter(chunk_size = 200, chunk_overlap = 20)
text_splitted = spiltter.split_documents([doc])
```

```python
text_splitted
```

```python
VDB = Chroma.from_documents(documents=text_splitted, embedding=embedding,persist_directory='gen-ai')
```
